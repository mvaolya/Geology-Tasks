```python
import pandas as pd
import numpy as np

# Последовательности типов пород
seq1 = taeznaya_102['Rock'].tolist()
seq2 = taeznaya_103['Rock'].tolist()

# Функция для вычисления расстояния Левенштейна
def levenshtein_distance(seq1, seq2):
    m = len(seq1)
    n = len(seq2)
    D = np.zeros((m+1, n+1), dtype=int)

    for i in range(m+1):
        D[i, 0] = i
    for j in range(n+1):
        D[0, j] = j

    for i in range(1, m+1):
        for j in range(1, n+1):
            cost = 0 if seq1[i-1] == seq2[j-1] else 1
            D[i, j] = min(
                D[i-1, j] + 1,    # Удаление
                D[i, j-1] + 1,    # Вставка
                D[i-1, j-1] + cost # Замена
            )
    return D

# Построение матрицы расстояний
distance_matrix = levenshtein_distance(seq1, seq2)
print("Матрица расстояний Левенштейна:")
print(distance_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_levenshtein_matrix(seq1, seq2, D):
    plt.figure(figsize=(10, 8))
    
    # Визуализация матрицы расстояний
    sns.heatmap(
        D, 
        annot=True, 
        fmt="d", 
        cmap="YlOrRd", 
        linewidths=0.5,
        xticklabels=[""] + seq2, 
        yticklabels=[""] + seq1,  
    )
    
    plt.title("Матрица расстояний Левенштейна", fontsize=14)
    plt.xlabel("Скважина 103 (Таёжное 103)")
    plt.ylabel("Скважина 102 (Таёжное 102)")
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

plot_levenshtein_matrix(seq1, seq2, distance_matrix)
```

```python
def get_edit_operations(description_102, description_103, D):
    i, j = len(description_102), len(description_103)
    operations = []

    while i > 0 or j > 0:
        if i > 0 and j > 0:
            cost = 0 if description_102[i-1] == description_103[j-1] else 1
            if D[i, j] == D[i-1, j-1] + cost:
                if cost == 0:
                    operations.append(('M', description_102[i-1], description_103[j-1]))  
                else:
                    operations.append(('R', description_102[i-1], description_103[j-1]))  
                i -= 1
                j -= 1
                continue
        if i > 0 and D[i, j] == D[i-1, j] + 1:
            operations.append(('D', description_102[i-1], '-'))  
            i -= 1
        elif j > 0 and D[i, j] == D[i, j-1] + 1:
            operations.append(('N', '-', description_103[j-1])) 
            j -= 1

    return operations[::-1]  

edit_operations = get_edit_operations(description_102, description_103, distance_matrix)

print("Корреляция:")
for op in edit_operations:
    print(f"{op[0]}: {op[1]} → {op[2]}")
```

```python
well_102 = pd.read_csv(r"C:\Users\Ольга\Downloads\litho\Таёжное 102.csv", sep=';')
well_103 = pd.read_csv(r"C:\Users\Ольга\Downloads\litho\Таёжное 103.csv", sep=';')

# Инвертируем порядок данных (чтобы глубина увеличивалась вниз)
well_102 = well_102.sort_values('Depth', ascending=False).reset_index(drop=True)
well_103 = well_103.sort_values('Depth', ascending=False).reset_index(drop=True)

# Цвета для типов пород
rock_colors = {
    'Угли': 'black',
    'Глины': 'green',
    'Песчаники крупнозернистые': 'yellow',
    'Известняки песчанистые': 'gray',
}

# Создаём рисунок
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 18))
plt.subplots_adjust(wspace=0.3)

# Функция для рисования скважины
def draw_well(ax, data, name, pos):
    min_depth = data['Depth'].min()
    max_depth = data['Depth'].max()
    current_depth = min_depth
    layers = []
    
    for i in range(len(data)):
        thickness = data.iloc[i]['H']
        rock_type = data.iloc[i]['Rock']
        bottom_depth = current_depth + thickness
        
        layers.append({
            'top': current_depth,
            'bottom': bottom_depth,
            'rock': rock_type,
            'center': (current_depth + bottom_depth)/2
        })
        
        # Рисуем пласт
        ax.barh(current_depth + thickness/2, 
               width=4, 
               height=thickness,
               left=pos,
               color=rock_colors.get(rock_type, 'lightgray'),
               edgecolor='black',
               linewidth=0.5)
        
        # Подписи глубин
        if i == 0:
            ax.text(pos + 4.5, current_depth, f'{current_depth:.1f} м', va='center')
        ax.text(pos + 4.5, bottom_depth, f'{bottom_depth:.1f} м', va='center')
        
        current_depth = bottom_depth
    
    ax.set_title(name, pad=20)
    ax.set_xlim(0, 10)
    ax.set_xticks([])
    ax.set_ylim(max_depth + 10, min_depth - 10)
    ax.grid(True, axis='y', linestyle=':', alpha=0.7)
    ax.set_ylabel('Глубина (м)', labelpad=15)
    
    return layers

# Рисуем скважины 
layers_102 = draw_well(ax1, well_102, "Скважина 102", 1)
layers_103 = draw_well(ax2, well_103, "Скважина 103", 6)

# Получаем последовательности пород
seq_102 = [layer['rock'] for layer in layers_102]
seq_103 = [layer['rock'] for layer in layers_103]

# Рисуем корреляционные линии для верхних и нижних границ 
for op in edit_operations:
    if op[0] == 'M':  # Только для совпадающих слоев
        i, j = op[3], op[4]
        layer1 = layers_102[i]
        layer2 = layers_103[j]
        
        # Параметры для линий
        line_props = {
            'arrowstyle': '-',
            'linewidth': 1.0,
            'linestyle': '--',
            'color': 'red',
            'alpha': 0.7
        }  
        # 1. Линия для верхних границ
        con_top = ConnectionPatch(
            xyA=(5, layer1['top']), xyB=(6, layer2['top']),
            coordsA="data", coordsB="data",
            axesA=ax1, axesB=ax2,
            **line_props
        )
        fig.add_artist(con_top)
        
        # 3. Линия для нижних границ
        con_bottom = ConnectionPatch(
            xyA=(5, layer1['bottom']), xyB=(6, layer2['bottom']),
            coordsA="data", coordsB="data",
            axesA=ax1, axesB=ax2,
            **line_props
        )
        fig.add_artist(con_bottom)
        
        # Подпись типа породы
        fig.text(0.5, (layer1['center'] + layer2['center'])/2 / (max(layers_102[-1]['bottom'], layers_103[-1]['bottom']) + 20),
                op[1], ha='center', va='center',
                bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))

# Легенда
legend_elements = [plt.Rectangle((0,0),1,1, color=c, label=l) 
                  for l,c in rock_colors.items() if l != 'Другие']
legend_elements.append(plt.Rectangle((0,0),1,1, color='lightgray', label='Другие'))
ax1.legend(handles=legend_elements, 
          loc='upper right',
          bbox_to_anchor=(1.9, 1),
          framealpha=1)

plt.tight_layout()
plt.show()
```
